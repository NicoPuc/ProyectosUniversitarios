# Pregunta 2
- A)<br>
 Primero X(0) es el vector que describe el numero de residentes que  compran cada uno de los productos. Entonces por la matriz dada de X(0) quiere decir que r1 tiene 10.000 residentes que compran el producto 1, r2 10.000 y r3 10.000, de la misma forma que r1.<br>

  PX(0): Estamos multiplicando la matriz P por el vector X(0), estariamos de alguna forma aplicando la matriz de los porsentajes a los distintos grupos de residentes, de esta forma estariamos desplazando el vector de tiempo 0 a tiempo 1, porque estamos aplicando la matriz una vez.<br>

  P(PX(0)) Como ya sabemos que PX(0) es equivalente a X(1) nos quedaria P(X(1)) que seria lo mismo que PX(1) = PPX(0) $=>$  P$^2$X(0), esto seria como tener la cantidad de recidentes que compran el producto en el mes 2. <br>

  P(P$^2$X(0)) Por lo demostrado recien sabemos que P$^2$X(0) = X(2) nos quedaria P(X(2)) => PX(2) => X(3) => P$^3$X(0). <br>

- B)  






```python
import numpy as np

P = np.array([
    [0.8, 0.2, 0.05],
    [0.05, 0.75, 0.05],
    [0.15, 0.05, 0.9]
])

#X(0)
X0 = np.array([10000, 10000, 10000])

# n numero de periodos, P es la matriz con los porcentajes y X0 el vector inicial
def X_n(n, P, X0):
    P_n = np.linalg.matrix_power(P, n)
    p_punto = np.dot(P_n, X0)
    return p_punto

valores = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
valores_finales = []
for i in valores:
  valor_temporal = X_n(i,P,X0)
  valores_finales.append(valor_temporal)

#imprimi asi para que queden con saltos de linea y quede mas claro los valores
for i in range(len(valores_finales)):
  print(f"{(i+1)*5}: {valores_finales[i]}")

```

El output nos entrega la informacion de la siguente forma:

n: [p1 p2 p3]
Siendo p el numero de residentes respecto a al producto que indique el subindice.Y n los valores a calcular de X(n). <br>

```python
import numpy as np
import matplotlib.pyplot as plt

P = np.array([
    [0.8, 0.2, 0.05],
    [0.05, 0.75, 0.05],
    [0.15, 0.05, 0.9]
])

X0 = np.array([10000, 10000, 10000])

#añadi el 0 porque me parecio interesante ver el comportamiento entre 0 y 5.
valores = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
valores_finales = []
for i in valores:
  valor_temporal = X_n(i,P,X0)
  valores_finales.append(valor_temporal)

# Utilizo generadores porque son ciclos muy simples
product_1 = [v[0] for v in valores_finales]
product_2 = [v[1] for v in valores_finales]
product_3 = [v[2] for v in valores_finales]

plt.figure(figsize=(10, 6))

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
# De ahi saque como usar .plot // la letra antes de la o indica el color la o que son circulos, s cuadrados y d diamantes y -- que es lina punteada
plt.plot(valores, product_1, 'go--', linewidth=2, markersize=12, label='Producto 1')
plt.plot(valores, product_2, 'bs--', linewidth=2, markersize=12, label='Producto 2')
plt.plot(valores, product_3, 'yd--', linewidth=2, markersize=12, label='Producto 3')

plt.xlabel('numero de meses', fontsize=10)
plt.ylabel('cantidad de residentes', fontsize=10)
plt.title('cantidad de residentes que comprar cada producto en el n-ésimo mes', fontsize=15)
plt.legend()

plt.show()
```
